```python
import os
from dotenv import load_dotenv
from supabase import create_client, Client

load_dotenv()

url: str = os.environ.get("SUPABASE_URL")
key: str = os.environ.get("SUPABASE_KEY")
supabase: Client = create_client(url, key)
```

```python
response = (
    supabase.rpc("nearby_places", {
        "lat": 107.07037164103,
        "long": 10.34982650293901,
    })
    .execute()
)
```

```python
res = supabase.rpc("fetch_places").execute()
res.data
```

```python
type(res.data)
```

```python
response = supabase.rpc(
    "upsert_place", {
        "p_name": "place_data.name",
        "p_tags": [],
        "p_latitude": 10.34982650293901,
        "p_longitude": 107.07037164103,
        "p_images": ["place_data.images"]
    }).execute()
```

```python
place = (
    supabase.table("places")
    .select("id")
    .eq("name", "place_data.name")
    .single()
    .execute()
)
```

```python
place.data
```

```python
place.data["id"]
```

```python
place_id = place.data["id"]

# Replace with the actual place ID

audio_guides = [
    {
        "title": "New Guide 1",
        "duration_seconds": 60,
        "audio_url": "https://example.com/audio1.mp3",
        "subtitle_url": "https://example.com/subtitle1.vtt"
    },
    {
        "title": "New Guide 2",
        "duration_seconds": 60,
        "audio_url": "https://example.com/audio2.mp3",
        "subtitle_url": "https://example.com/subtitle2.vtt"
    }]
# Call the update_audio_guides function

response = supabase.rpc("update_audio_guides",
                        {
                            "p_place_id": place_id,
                            "audio_guides": audio_guides
                        }).execute()
```
