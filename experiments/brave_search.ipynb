```python
import os
import requests
import json
from dotenv import load_dotenv

load_dotenv()
```

```python
res = requests.get("https://api.search.brave.com/res/v1/web/search",
                   headers={
                       "X-Subscription-Token": os.getenv("BRAVE_API_KEY"),
                       "Accept": "application/json",
                       "Accept-Encoding": "gzip"
                   },
                   params={
                       "q": "Các địa điểm du lịch văn hóa Việt Nam",
                       "count": 10
                   })

print(json.dumps(res.json(), indent=2, ensure_ascii=False))
```

```python
from pydantic import BaseModel

class SearchResult(BaseModel):
    title: str
    url: str
    description: str
    page_age: str
    thumbnail_original: str
    thumbnail_src: str

    def __init__(self, json_data: dict):
        super().__init__(
            title = json_data["title"],
            url = json_data["url"],
            description = json_data["description"],
            page_age = json_data["page_age"] if "page_age" in json_data else "",
            thumbnail_original = json_data["thumbnail"]["original"] if "thumbnail" in json_data else "",
            thumbnail_src = json_data["thumbnail"]["src"] if "thumbnail" in json_data else "",
        )
```

```python
search_results = []

for r in res.json()["web"]["results"]:
    search_results.append(SearchResult(r))
```
