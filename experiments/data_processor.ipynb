```python
import openai
import json
import os
from dotenv import load_dotenv
from rich.markdown import Markdown
from typing import List, Optional
from pydantic import BaseModel

load_dotenv()
```

```python
# client = openai.OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")
client = openai.AzureOpenAI(
    api_version=os.environ.get("AOAI_API_VERSION"),
    azure_endpoint=os.environ.get("AOAI_ENDPOINT"),
)

class Place(BaseModel):
    name: str
    description: str
    image_url: Optional[str]
    address: str
    phone: Optional[str]
    website: Optional[str]
    opening_hours: Optional[str]
    rating: Optional[float]
    pros: list[str]
    cons: list[str]
    
class PlaceList(BaseModel):
    places: List[Place]
```

```python
place_data = open("data/TOPLIST 23 địa điểm du lịch Sài Gòn nghe là muốn xách balo lên và đi.txt", "r", encoding="utf-8").read()

prompt = f"""Content:
    {place_data}
---
You are a travel guide. 
Your task is to go through each of the places in the given content and extract information about each place.

Please provide the information of each place in JSON format.
If any of the information is unknown, assign null value instead.
"""


completion = client.beta.chat.completions.parse(
    temperature=0.8,
    model="lg-gpt-4o-mini",
    messages=[
        {"role": "user", "content": prompt},
    ],
    response_format=PlaceList,
)
```

```python
json_data = json.loads(completion.choices[0].message.content)
print(json.dumps(json_data, indent=2, ensure_ascii=False))
```
